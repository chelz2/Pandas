```python
# Wikipedia --> we'll read html tables from wikipedia article webpage --> World population

import pandas as pd
```

```python
# pandas ---> grabs all tables in webpage from which we can select table we are intersested in

url = "https://en.wikipedia.org/wiki/World_population"  # ---> url world population wikipedia 
```

```python
tables = pd.read_html(url)  # ---> pandas grabs all the html table in a list

#tables  # ---> returns list of tables from webpage
```

```python
len(tables)  # ---> retuurns number of tables in list 
```

```python
tables[0]  #---> index at 0, returns the first table from list of 30 tables as dataframes
```

```python
tables[29] #---> last html table as df
```

```python
tables[3]
```

```python
tables[4]
```

```python
# multi index in html tables ---> i.e pandas read higher level table above table header in html table as multi index

tables[4].columns  #---> returns column header & we can see if table is multi index 
```

```python
# cleaning html table datas 

top_ten = tables[4]
```

```python
top_ten
```

```python
top_ten = top_ten.drop('% of world',axis=1) # ---> we can drop % world column since it is NULL
```

```python
top_ten
```

```python
top_ten.columns = ['country', 'population', 'date', 'source'] # ---> re-assigning column names
```

```python
top_ten
```

```python
# NB --> i.e once we grabs html tables from webpage url or raw-html files  as pandas dataframes, we can do all data cleaning & manipulation that we can with pandas :)
```

```python
# save / export pandas df as html tables in thml files

top_ten.to_html('topten.html', index=False)  # ----> saves topten df as html table in new html call topten.html
```

```python
df = pd.read_html('topten.html')  # ---> reading the new topten.html we saved 
```
